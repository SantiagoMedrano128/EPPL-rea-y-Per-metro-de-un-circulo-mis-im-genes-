## Formula para calcular area

## Formula para calcular el perimetro
# $P=2*π*r $

```python
# Calcular area y perimetro de un circulo
# Variables A, P, r
# Formulas A = pi * r^2 y P = 2 * pi * r
import math
print("Calcular area y perimetro de un circulo")
r = float(input("Por favor, introduce el valor del radio (r): "))
A = math.pi * (r**2)
P = 2 * math.pi * r
print(f"El area del circulo es: {A}")
print(f"El perimetro del circulo es: {P}")
print("Santiago Medrano NL = 37")
```
